## Feature engineering

### New variables

##### jitter-shimmer-difference ratio = Jitter(d.d.p) / Shimmer(d.d.a)

Both Jitter and Shimmer are pivotal for voice quality. So, we choose "jitter_shimmer_difference_ratio" because it combines variations between consecutive voice cycles, providing a comprehensive assessment of voice stability and instability, particularly vital in distinguishing Parkinson's symptoms.

AZADI, H., AKBARZADEH-T, M.-R., SHOEIBI, A. & KOBRAVI, H. R. 2021. Evaluating the Effect of Parkinson's Disease on Jitter and Shimmer Speech Features. Advanced biomedical research [Online], 10. Available: http://europepmc.org/abstract/MED/35127581
https://doi.org/10.4103/abr.abr_254_21
https://europepmc.org/articles/PMC8781904
https://europepmc.org/articles/PMC8781904?pdf=render [Accessed 2021].


```python
jitter_shimmer_difference_ratio = df.iloc[:,5]/df.iloc[:,11]
jitter_shimmer_difference_ratio_df = pd.DataFrame(jitter_shimmer_difference_ratio)
jitter_shimmer_difference_ratio_df
```

##### Harmonicity-to-Jitter ratio = Harmonic-to-Noise ratio(HNR) / Jitter(%)

Harmonicity-to-Jitter ratio (HJR) might be a good combonation because it captures the essence of the disease's effect on voice: a likely increase in jitter (frequency instability) and a decrease in harmonicity (voice clarity and quality). The balance (or imbalance) between these two factors, as captured by the HJR, could be a distinctive signature of PD.

FERRAND, C. T. 2002. Harmonics-to-noise ratio: an index of vocal aging. Journal of voice, 16, 480-487.

FARRÃšS, M., HERNANDO, J. & EJARQUE, P. Jitter and shimmer measurements for speaker recognition.  8th Annual Conference of the International Speech Communication Association; 2007 Aug. 27-31; Antwerp (Belgium).[place unknown]: ISCA; 2007. p. 778-81., 2007. International Speech Communication Association (ISCA).

```python
harmonicity_to_jitter_ratio = df.iloc[:,14]/df.iloc[:,1]
harmonicity_to_jitter_ratio_df = pd.DataFrame(harmonicity_to_jitter_ratio)
harmonicity_to_jitter_ratio_df
```

#####  Pitch range = Maximum pitch - Minimum pitch

This is the difference between Maximum pitch and Minimum pitch. Reduced pitch range (monotone) is one of the most common speech problems experienced by people with Parkinsonâ€™s disease. This means that people with PD may have difficulty conveying emotions when they speak, as their voice may sound flat and unexpressive. In contrast, healthy individuals typically have a wider pitch range, allowing them to convey emotions more effectively through their voice.

RAMIG, L. O., FOX, C. & SAPIR, S. 2008. Speech treatment for Parkinsonâ€™s disease. Expert Review of Neurotherapeutics, 8, 297-309.

```python
pitch_range = df.iloc[:,19] - df.iloc[:,18]
pitch_range_df = pd.DataFrame(pitch_range)
pitch_range_df
```

##### Pulse-to-Pitch ratio = Standard deviation of period / Standard deviation of pitch

The main reason to choose the ratio of the standard deviation of period to the standard deviation of pitch to distinguish people with Parkinson's Disease (PD) from healthy individuals is that Parkinson's affects the muscles used in speech, leading to vocal abnormalities. This ratio captures the variability in both the duration of the vocal fold vibration and the fundamental frequency, providing a combined metric that may sensitively detect these voice abnormalities characteristic of PD.

KHAN, T., WESTIN, J. & DOUGHERTY, M. 2014. Classification of speech intelligibility in Parkinson's disease. Biocybernetics and Biomedical Engineering, 34, 35-45.

```python
pulse_to_pitch_ratio  = df.iloc[:,23] / df.iloc[:,17]
pulse_to_pitch_ratio_df = pd.DataFrame(pulse_to_pitch_ratio)
pulse_to_pitch_ratio_df
```

### Add new variables to make a new dataset

```python
# Combine the new variables with the old data to create a single dataset, getting it ready for analysis.
new_df = pd.concat((df.iloc[:,0:28],
                    jitter_shimmer_difference_ratio_df,
                    harmonicity_to_jitter_ratio_df,
                    pitch_range_df,
                    pulse_to_pitch_ratio_df,
                    df.iloc[:,28]), axis=1, ignore_index=True)

new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# Save the new data to a new .txt file
new_df.to_csv('new_po1_data.txt', index=False, header=False)
```
